## Dataset Structure and General Overview

The processed dataset for this study contains 359 samples and 2017 columns. The structure of the data is as follows:

1. **Environmental Variables:**  
  The first two columns are continuous features:
     - `Avg_Temp_C` (Average Temperature, °C)
     - `CO2_Emission_Mt` (CO₂ Emissions in Megatonnes)

2. **Genomic Sequence Variables:**  
  The remaining 2015 columns (`p_1` to `p_2015`) represent position-wise nucleotide data from the targeted DENV genomic region. Each position is encoded as:
    - `1`: Adenine (A)
    - `2`: Uracil (U) / Thymine (T)
    - `3`: Guanine (G)
    - `4`: Cytosine (C)
    - `0`: Unknown or Padding (N)

### Summary Statistics and Data Types

- All columns are numeric: the environmental variables are `float64`, and the sequence positions are `int64`. This is optimal for efficient downstream analysis and modeling.
- Each column contains 359 entries, confirming there are no missing values in any feature.
- The memory footprint is modest (5.5 MB), supporting interactive exploration.

**Environmental Features:**  
  - Average temperature values (`Avg_Temp_C`) range from approximately 24.0°C to 26.2°C, with a mean around 25.7°C.
  - CO₂ emissions (`CO2_Emission_Mt`) vary between 147.1 Mt and 2489.9 Mt, with a mean of about 2181.8 Mt.
  - The distribution of both variables appears continuous and suitable for binning or direct modeling.

**Sequence Features:**  
  - Each nucleotide position (`p_1` to `p_2015`) has a full set of 359 values, though the min and lower quartile are frequently zero, reflecting substantial padding/unknowns in the dataset.
  - Means and standard deviations of sequence columns vary, indicating heterogeneous sequence composition across the genome.
  - Max values reach `4` for all sequence columns, confirming the presence of all possible base codes.
  - The prevalence of zeros at many positions suggests some sequences are shorter or have regions that were padded to the full length, which is typical in viral sequence analysis.

#### Overall Assessment

- **Data completeness** is high, with no missing entries in any variable.
- **Environmental variables** provide reasonable dynamic range and variance for modeling and association studies.
- **Sequence data** shows expected heterogeneity for a viral dataset, but padding prevalence must be considered in downstream analyses and interpretation.

### Shannon Entropy per Position 

```python
def shannon_entropy(arr):
    values, counts = np.unique(arr, return_counts=True)
    probs = counts / counts.sum()
    return entropy(probs, base=2)

entropies = [shannon_entropy(df[col].values) for col in seq_cols]

plt.figure(figsize=(14,4))
plt.plot(range(1, len(seq_cols)+1), entropies)
plt.xlabel("Sequence Position")
plt.ylabel("Shannon Entropy")
plt.title("Shannon Entropy by Position")
plt.show()
```

### Shannon Entropy by Sequence Position

This plot visualizes the Shannon entropy at each sequence position, quantifying the uncertainty or diversity of nucleotide usage at that site. Higher entropy indicates more evenly distributed bases (greater variability), while lower entropy reflects more conservation.

#### Key Observations
- **Broadly high entropy across most positions**:
The majority of sequence positions maintain an entropy around 1.7–2.0, consistent with high diversity (close to the theoretical maximum of ~2.32 for five possible values: A, T, G, C, N).
This matches previous findings that all four canonical bases (and sometimes 'N') are well represented at nearly every position.

- **Entropy increases further after position ~1750**:
The last several hundred positions show a spike in entropy, exceeding 2.2, before rapidly dropping near the very end.
This spike is likely due to a mixture of nucleotides and a sharp increase in the proportion of padding/unknowns ('N'), as seen in the padding analysis. The artificial mixing of valid bases with high rates of 'N' can inflate entropy estimates.

- **Sharp decline in entropy at the very end**:
The most distal positions show very low entropy, reflecting that nearly all values here are 'N'/padding (very little diversity).

#### Overall Assessment
- The region from positions 1 to ~1750 is highly variable and data-rich, with little evidence of conserved (low entropy) sites.
- The rising entropy at the end reflects increased uncertainty from padding/unknowns and should not be interpreted as true biological diversity.
- Positions with extreme entropy values (either very low or very high, especially at the tails) may not be reliable for association studies or modeling.

## Genotype-Environment Association

### Define Binning for Environmental Variables

```python
# Define temperature and CO2 bins using quartiles
df['Temp_Bin'] = pd.qcut(df['Avg_Temp_C'], q=4, labels=['Q1 (Coolest)', 'Q2', 'Q3', 'Q4 (Hottest)'])
df['CO2_Bin'] = pd.qcut(df['CO2_Emission_Mt'], q=4, labels=['Q1 (Lowest)', 'Q2', 'Q3', 'Q4 (Highest)'])
```

```python
# Check distribution
print("Temperature bins:")
print(df['Temp_Bin'].value_counts())
print("\nCO2 bins:")
print(df['CO2_Bin'].value_counts())
```

### Binning of Environmental Variables

To explore possible associations between environmental conditions and genotype patterns, the continuous environmental variables (average temperature and CO₂ emissions) were divided into discrete quartile-based bins. This stratification enables categorical comparison between different environmental regimes.

#### Overall Assessment
- The binning is roughly balanced, although Q1 (Lowest CO₂) and Q1 (Coolest Temp) have the largest groups, reflecting data skew toward certain environmental conditions (recent years with similar values).
- Having all four quartile bins represented ensures that comparisons will capture both low and high ends of the environmental spectrum.
- This approach provides a foundation for subsequent genotype-environment association analyses, such as chi-squared and Cramér’s V, by allowing the examination of nucleotide distributions at each position within discrete environmental strata.

### Association Scanning Function (Cramér’s V and Chi-squared p-value)

```python
def scan_positions_for_association(df, seq_cols, bin_col):
    results = []
    for pos in seq_cols:
        contingency = pd.crosstab(df[pos], df[bin_col])
        if contingency.shape[0] > 1:  # Only if more than one nucleotide observed
            chi2, p, dof, expected = chi2_contingency(contingency)
            n = contingency.sum().sum()
            cramers_v = np.sqrt(chi2 / (n * (min(contingency.shape)-1)))
            results.append({'Position': pos, 'p_value': p, 'CramersV': cramers_v})
    assoc_df = pd.DataFrame(results)
    return assoc_df.sort_values('CramersV', ascending=False).reset_index(drop=True)
```

### Run Association Scan for Temperature

```python
# Get sequence column names
seq_cols = [col for col in df.columns if col.startswith('p_')]

# Run association scan for temperature bins
assoc_temp = scan_positions_for_association(df, seq_cols, 'Temp_Bin')

print("Top 10 positions associated with temperature:")
display(assoc_temp.head(10))
```

### FDR Correction for Temperature

```python
# Apply FDR correction (Benjamini-Hochberg)
reject, pvals_corrected, _, _ = multipletests(assoc_temp['p_value'], method='fdr_bh')
assoc_temp['FDR_p'] = pvals_corrected
assoc_temp['Significant'] = reject

# Show significant positions after FDR
print("Significant positions for temperature after FDR correction:")
display(assoc_temp[assoc_temp['Significant']].head(10))
```

### Genotype–Temperature Association Analysis

To identify specific sequence positions whose nucleotide distribution shows a significant association with average temperature bins.

#### Overall Assessment

- The presence of several strongly associated positions (even after FDR correction) indicates that certain nucleotide sites in the DENV genome may be influenced by environmental temperature.
- These positions are candidates for further investigation — for example, mapping them to known functional regions of the genome (e.g., E protein, UTRs) or looking for overlap with positions under known selective pressure.
- Moderate effect sizes (Cramér’s V < 0.3) are common in biological data and suggest a real but not overwhelming influence; environmental temperature may contribute to the observed nucleotide variation alongside other evolutionary forces.

### Run Association Scan for CO₂

```python
# Run association scan for CO2 bins
assoc_co2 = scan_positions_for_association(df, seq_cols, 'CO2_Bin')

print("Top 10 positions associated with CO2:")
display(assoc_co2.head(10))
```

### FDR Correction for CO₂

```python
# Apply FDR correction for CO2 associations
reject_co2, pvals_corrected_co2, _, _ = multipletests(assoc_co2['p_value'], method='fdr_bh')
assoc_co2['FDR_p'] = pvals_corrected_co2
assoc_co2['Significant'] = reject_co2

# Show significant positions after FDR for CO2
print("Significant positions for CO2 after FDR correction:")
display(assoc_co2[assoc_co2['Significant']].head(10))
```

### Genotype–CO₂ Association Analysis

To identify sequence positions whose nucleotide composition shows significant association with CO₂ emission bins, indicating possible environmental influence on viral genotype.

#### Overall Assessment

- The presence of several sequence positions with extremely low p-values and moderate effect sizes suggests that CO₂ emission levels may be associated with specific patterns of nucleotide variation in the viral genome.
- Some positions (e.g., p_17, p_1839) appear in both the temperature and CO₂ association scans, indicating they may be especially sensitive to broad environmental changes.
- These findings support the hypothesis that environmental factors, including CO₂, may exert evolutionary pressure on the dengue virus genome.

### Visualize Cramér’s V for Both Variables 

```python
plt.figure(figsize=(14, 4))
plt.plot(assoc_temp['CramersV'].values, label='Temperature')
plt.plot(assoc_co2['CramersV'].values, label='CO2')
plt.xlabel('Ranked Sequence Positions')
plt.ylabel("Cramér's V")
plt.title("Cramér's V Association Across All Sequence Positions")
plt.legend()
plt.show()
```

### Cramér’s V Association Across All Sequence Positions
 
To globally compare the strength of association (Cramér’s V) between sequence positions and environmental variables (Temperature and CO₂), revealing the landscape of genotype–environment associations across the viral genome.

#### Key Insights

- **General Pattern:**  
  The plot shows the Cramér’s V values for all sequence positions, ranked from highest to lowest, for both temperature and CO₂ bins.
  - The orange line (CO₂) consistently lies above the blue line (Temperature) across nearly all positions.
  - This indicates that, genome-wide, CO₂ emission bins explain a larger fraction of the variance in base composition at each position than temperature bins do.

- **Magnitude of Association:**  
  - The highest Cramér’s V values for both variables are in the moderate range (~0.29 for temperature, ~0.34 for CO₂).
  - These effect sizes suggest moderate but widespread associations rather than a few isolated, strongly environment-dependent positions.

- **Genome-Wide Trend:**  
  - The gradual decline in both curves (rather than a sharp drop) shows that a large number of positions have at least some environmental association—supporting the hypothesis that environmental factors might exert broad, polygenic pressure on the dengue virus genome.

- **Comparative Conclusion:**  
  - CO₂ levels appear to be a stronger environmental correlate of viral genomic variation than temperature within this dataset and region.
  - This could reflect either direct biological effects of CO₂ (e.g., as a proxy for urbanization or host/vector density), or indirect correlation with other unmeasured environmental/epidemiological variables.
